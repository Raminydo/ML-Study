```python
import pandas as pd
import matplotlib.pyplot as plt
from sklearn.linear_model import LinearRegression
from sklearn.metrics import mean_squared_error, mean_absolute_error, root_mean_squared_error
```

```python
# Dataset: https://www.kaggle.com/datasets/prokshitha/home-value-insights
dataset = pd.read_csv('house_price.csv')
dataset.head()
```

```python
dataset.isna().sum()
```

```python
X = dataset.drop('House_Price', axis=1)
y = dataset['House_Price']
```

```python
model = LinearRegression()
model.fit(X, y)
```

```python
y_pred = model.predict(X)
```

```python
mse = mean_squared_error(y, y_pred)
mae = mean_absolute_error(y, y_pred)
rmse = root_mean_squared_error(y, y_pred)
```

```python
print(f'MSE: {mse}')
print(f'MAE: {mae}')
print(f'RMSE: {rmse}')
```

```python
plt.plot(y)
plt.plot(y_pred)
plt.show()
```

```python
print('a:\n', model.coef_)
print('\nb:\n', model.intercept_)
```
