```python
from sklearn.datasets import make_moons, make_circles
import matplotlib.pyplot as plt
import numpy as np
from sklearn.metrics import silhouette_score
from sklearn.cluster import DBSCAN, KMeans
from sklearn.preprocessing import StandardScaler
```

# moon

```python
X, y = make_moons(n_samples=100, noise=0.1)
plt.scatter(X[:, 0], X[:, 1], c=y)
```

```python
plt.scatter(X[:, 0], X[:, 1])
```

```python
scaler = StandardScaler()
X_scaled = scaler.fit_transform(X)
```

```python
model = DBSCAN(eps=0.6, min_samples=9)
label = model.fit_predict(X_scaled)

plt.scatter(X[:, 0], X[:, 1], c=label)
```

```python
plt.scatter(X[:, 0], X[:, 1], c=label)

outlier = X[label == -1]
plt.scatter(outlier[:, 0], outlier[:, 1], c='red', edgecolors='red', s=100, alpha=0.3)

core_points = X[model.core_sample_indices_]
plt.scatter(core_points[:, 0], core_points[:, 1], c='green', edgecolors='green', s=100, alpha=0.3)

plt.show()
```
